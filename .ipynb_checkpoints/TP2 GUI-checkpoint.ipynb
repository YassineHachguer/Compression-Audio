```python
import PySimpleGUI as sg
import os.path


_VARS = {'window': False,
         'fig_agg': False,
         'pltFig': False}

    


full_canvas = False
    

#left-hand column elements
file_list_column = [
    [sg.Text('Selectionner un dossier: ')],
    [
        sg.Text("Dossier:"),
        sg.In(size=(19, 1), enable_events = True, key="-FOLDER-"),
        sg.FolderBrowse(),
    ],
    [
        sg.Listbox(
            values=[],enable_events=True, size=(35, 25), key="-FILE LIST-"
        )
    ],
]

#centre column elements
visual_column = [
    [sg.Button("Reconstituer L'image", key = '-IMAGE REC-', size = (18,1), disabled = True)],
    [sg.Image('place_holder.png',key="-IMAGE-")],
    [sg.Button("Reconstituer L'audio", key = '-AUDIO REC-', size = (18,1), disabled = True)],
    [sg.Button("play", key = "-PLAY-", disabled = True), sg.Button("stop", key = "-STOP-", disabled = True)],
    [sg.Button("Reconstituer Le text", key = '-TEXT REC-', size = (18,1), disabled = True)],
    [sg.Text("Votre fichier text...", size=(40,10), background_color="#F0F3F7", text_color="black", key='-TEXT-')],

]

Codages = ["RZ","NRZ","NRZI", "Manchester", "Manchester diff√©rentiel", "Miller", "CMI",
"AMI", "MLT3", "2B1Q"]

#right-hand column elements
button_list_column = [
    [sg.Text("Selectionner un codage en ligne:")],
    [sg.Combo(Codages,size=(22,1) ,  expand_x=True, enable_events=True,  readonly=False, key='-COMBO-')],
    [sg.Button("Afficher binaire", key = "-SELECT-", size=(15,1))],
    [sg.Text("Chaine binaire...", size=(35,2), background_color="#F0F3F7", text_color="black", key="-BINARY SEQUENCE-")],
    [sg.Button("Tracer le signal", key= "-PLOT-",size=(15,1), disabled = True)],
    [sg.Canvas(size = (30,20) ,key = "-CANVAS-")],
]

layout = [
    [
        sg.Column(file_list_column),
        sg.VSeperator(),
        sg.Column(visual_column),
        sg.VSeperator(),
        sg.Column(button_list_column),
    ]
]

_VARS['window'] = sg.Window("Codages en ligne", layout, finalize = True)



def draw_figure(canvas, figure):
    figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)
    figure_canvas_agg.draw()
    figure_canvas_agg.get_tk_widget().pack(side='top', fill='both', expand=1)
    return figure_canvas_agg


def draw_chart():
    _VARS['pltFig'] = plt.figure(figsize=(4,3))
    plt.plot(cdg.CEL(binary_sequence[:10], 100, 5, codage))
    _VARS['fig_agg'] = draw_figure(_VARS['window']['-CANVAS-'].TKCanvas, _VARS['pltFig'])
    
def update_chart():
    _VARS['fig_agg'].get_tk_widget().forget()
    plt.clf()
    plt.plot(cdg.CEL(binary_sequence[:10], 100, 5, codage))
    _VARS['fig_agg'] = draw_figure(_VARS['window']['-CANVAS-'].TKCanvas, _VARS['pltFig'])



while True:
    
    event, values = _VARS['window'].read(timeout=200)
    if event == "Exit" or event == sg.WIN_CLOSED:
        break
    if event == "-FOLDER-":
        folder = values["-FOLDER-"]
        try:
            # Get list of files in folder
            file_list = os.listdir(folder)
        except:
            file_list = []

        fnames = [
            f
            for f in file_list
            if os.path.isfile(os.path.join(folder, f))
        ]
        _VARS['window']["-FILE LIST-"].update(fnames)
    elif event == "-FILE LIST-":  # A file was chosen from the listbox
        try:
            _VARS['window']["-PLAY-"].update(disabled=True)
            _VARS['window']["-STOP-"].update(disabled=True)
            filename = os.path.join(
                values["-FOLDER-"], values["-FILE LIST-"][0]
            )

        except:
            pass
    if event == "-SELECT-":
        try:
            if filename.lower().endswith(('.txt')):
                binary_sequence = cdg.textfile2binary(filename)
                _VARS['window']['-TEXT REC-'].update(disabled = False)
                _VARS['window']['-IMAGE REC-'].update(disabled = True)
                _VARS['window']['-AUDIO REC-'].update(disabled = True)
                
                
            elif filename.lower().endswith(('.png','.jpeg','.jpg','.webp','.tiff','.tif','.ico','.bmp')):
                image = Image.open(filename)
                binary_sequence = cdg.image2binary(image)
                _VARS['window']['-TEXT REC-'].update(disabled = True)
                _VARS['window']['-IMAGE REC-'].update(disabled = False)
                _VARS['window']['-AUDIO REC-'].update(disabled = True)
            elif filename.lower().endswith(('.mp3','.wav','.wma','.au','.flac','.m4a','.mp4','.ogg', '.aiff','.aif','.aifc')):
                audio = AudioSegment.from_file(file=filename)
                fs = audio.frame_rate
                Nc = audio.channels
                binary_sequence = cdg.audio2binary(filename)
                _VARS['window']['-TEXT REC-'].update(disabled = True)
                _VARS['window']['-IMAGE REC-'].update(disabled = True)
                _VARS['window']['-AUDIO REC-'].update(disabled = False)
            else :
                binary_sequence = ''
                
            _VARS['window']['-BINARY SEQUENCE-'].update(binary_sequence[:50])
            
            _VARS['window']['-PLOT-'].update(disabled = False)    
        except Exception as e:
            print(e)
        
    if event == "-PLOT-":
        try:
            codage = values["-COMBO-"]
            if full_canvas == False:
                draw_chart()
                full_canvas = True
            else: 
                update_chart()
        except Exception as e:
            print(e)
    if event == "-TEXT REC-" and filename.lower().endswith(('.txt')):
        try:
            cdg.binary2textfile(binary_sequence, 'new_file.txt')
            with open('new_file.txt') as text_file:
                content = text_file.read()
            _VARS['window']['-TEXT-'].update(content)
        except Exception as e:
            print(e) 
            pass
    if event == "-IMAGE REC-" and filename.lower().endswith(('.png','.jpeg','.jpg','.webp','.tiff','.tif','.ico','.bmp')):
        try:
            
            rec_image = Image.fromarray(cdg.binary2image(binary_sequence))
            with BytesIO() as bio:
                rec_image.resize((256,rec_image.size[1]*256//rec_image.size[0])).save(bio, format="PNG")
                image_bytes = bio.getvalue()
            
            _VARS['window']["-IMAGE-"].update(data=image_bytes)
        except Exception as e:
            print(e)
    if event == "-AUDIO REC-" and filename.lower().endswith(('.mp3','.wav','.wma','.au','.flac','.m4a','.mp4','.ogg', '.aiff','.aif','.aifc')):
        try:
            rec_audio = cdg.binary2audio(binary_sequence)
            _VARS['window']["-PLAY-"].update(disabled=False)
            _VARS['window']["-STOP-"].update(disabled=False)
        except Exception as e:
            print(e)
            
    if event == "-PLAY-":
        try:
            if Nc == 1:
                sd.play(rec_audio, fs)
            elif Nc == 2:
                rec_audio.shape = -1, 2
                sd.play(rec_audio, fs)
            
        except Exception as e:
            print(e)

        
    if event == "-STOP-":
        try:
            sd.stop(rec_audio)
        except Exception as e:
            pass

        
        

            
        
_VARS['window'].close()
```

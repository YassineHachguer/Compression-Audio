```python
import Innovation
from tkinter import *
from tkinter import scrolledtext
import tkinter as tk
from tkinter import filedialog
from PIL import ImageTk, Image
import numpy as np
import matplotlib.pyplot as plt
import wave
import sounddevice as sd
import pydub
from pydub.playback import play
from pydub import AudioSegment
import ffmpeg
import os
import soundfile 
import wave
import scipy.signal as signal
import numpy as np
import sounddevice as sd
import struct
import audioop
from pydub.utils import get_array_type
from pydub import AudioSegment
from tkinter import ttk
```

```python
# Fonction pour sélectionner un fichier
def select_fichier_avant():
    global nom_fichier
    nom_fichier = filedialog.askopenfilename()
#Fonction pour selectionner un fichier compressée
def select_fichier_après():
    global nom_fichier_comp
    nom_fichier_comp = filedialog.askopenfilename()
# Fonction pour lire le fichier binaire sélectionné
def lire_info_fichier_selectionne():
    global nom_fichier
    from Innovation import lire_info
    contenu_info = lire_info(nom_fichier)
    # Afficher le contenu binaire dans la zone de texte
    zone_texte.config(state='normal')
    zone_texte.delete('1.0', tk.END)
    zone_texte.insert(tk.END, contenu_info)
    zone_texte.config(state='disabled')
def ecouter_audio():
    global nom_fichier
    from Innovation import ecouter
    ecouter(nom_fichier)
def collecte_info():
    global nom_fichier
    global nom_fichier_comp
    global output_file
    from Innovation import collecter_info_compression
    colect=collecter_info_compression(nom_fichier,output_file)
    # Afficher le contenu dans la zone de texte
    zone_texte.config(state='normal')
    zone_texte.delete('1.0', tk.END)
    zone_texte.insert(tk.END, colect)
    zone_texte.config(state='disabled')
def compresser_file():
    from Innovation import compresser_audio
    global nom_fichier
    global output_file
    # Récupération du fichier d'entrée sélectionné par l'utilisateur
    input_file = nom_fichier

    # Récupération du format de compression entré par l'utilisateur
    formatc = code_combobox.get()

    # Génération du nom de fichier de sortie
    output_file = input_file.split('.')[0] + '_compressed.' + formatc

    # Compression du fichier audio
    compresser_audio(input_file, output_file, formatc)
    #afficher un message
    zone_texte.config(state='normal')
    zone_texte.delete('1.0', tk.END)
    zone_texte.insert(tk.END, 'Votre compression a été effectuée avec succès.')
    zone_texte.config(state='disabled')
def decompresser_file():
    from Innovation import decompresser_audio_file
    global nom_fichier
    # Récupération du fichier d'entrée sélectionné par l'utilisateur
    input_file = nom_fichier

    # Récupération du format de compression entré par l'utilisateur
    formatc = format_entry.get()

    # Génération du nom de fichier de sortie
    output_file = input_file.split('.')[0] + '_decompressed.' + formatc

    # Compression du fichier audio
    decompresser_audio_file(input_file, output_file, formatc)
    #afficher un message
    zone_texte.config(state='normal')
    zone_texte.delete('1.0', tk.END)
    zone_texte.insert(tk.END, 'Votre decompression a été effectuée avec succès.')
    zone_texte.config(state='disabled')
#--------------------------------------------------Partie Inovation-----------------------------------------------------------#
def open_file_():
    global filename
    global raw_data
    from Innovation import open_file
    filename = filedialog.askopenfilename()
    raw_data = open_file(filename)

def writing_audio_irm_():
    global raw_data
    from Innovation import writing_audio_irm
    writing_audio_irm(raw_data)
    
def open_file_irm():
    global filename_irm
    global recovored_audio_data
    global sample_width
    global n_channels
    global sample_rate
    from Innovation import reading_audio_irm
    filename_irm = filedialog.askopenfilename()
    recovored_audio_data , sample_width, n_channels, sample_rate = reading_audio_irm(filename_irm)

def listen_():
    global recovored_audio_data
    global n_channels
    global sample_rate
    from Innovation import listen
    print(n_channels)
    listen(recovored_audio_data, n_channels, sample_rate)
```
